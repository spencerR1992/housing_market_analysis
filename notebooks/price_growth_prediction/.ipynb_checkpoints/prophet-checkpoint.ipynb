```python
import pandas as pd
from tqdm import tqdm
from prophet import Prophet
from multiprocessing import Pool
```

```python




df = pd.read_csv("../../data/Neighborhood_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
df.head()
```

```python
print(len(df))
print(df.columns)
```

```python
'''
Let's do the following: 
Train an autoregressive model on this data and predict the housing price growth for the next 5 years. 

'''
```

```python
# let's transform this data
nd_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName',
       'State', 'City', 'Metro', 'CountyName']
date_cols = [x for x in df.columns if x not in nd_cols]

```

```python
df = df[df.State=='TX']
df = df[df.Metro.isin(['Austin-Round Rock', 'San Antonio-New Braunfels'])]

df.head()
```

```python
dat = []


def get_pred(r):
    z = {}
    for c in nd_cols:
        z[c] = r[c]
    d = pd.DataFrame(data = {"ds": date_cols, "y": [r[d] for d in date_cols]})
    m = Prophet()
    _ = m.fit(d)
    future = m.make_future_dataframe(periods = 120, freq = "M")
    forecast = m.predict(future)
    z['df'] = forecast
    return z

with Pool(7) as p:
    r = list(tqdm(p.imap(get_pred, df.to_dict(orient = 'rows')), total = len(df)))
```

```python
# need to get summary stats for each of these dataframes. 
exdf = r[0]['df']
```

```python
exdf = exdf.copy()
```

```python
exdf = exdf.set_index("ds")
```

```python
exdf.head()
```

```python
def get_key_stats(df):
    d = {}
    # 12mo change
    zero = df.loc["2022-02-28"]['yhat']
    m_12 = df.loc["2021-02-28"]['yhat']
    m_24 =df.loc["2020-02-29"]['yhat']
    m_36 =df.loc["2019-02-28"]['yhat']
    m_48=df.loc["2018-02-28"]['yhat']
    d["12mo"] = (zero - m_12)/m_12
    d['24mo'] = (zero-m_24)/m_24
    d['36mo'] = (zero-m_36)/m_36
    d['48mo'] = (zero-m_48)/m_48
    
    
    
    p_12 = df.loc["2023-02-28"]['yhat']
    p_24 = df.loc['2024-02-29']['yhat']
    p_36 = df.loc['2025-02-28']['yhat']
    p_48 = df.loc['2026-02-28']['yhat']
    d['+12mo'] = (p_12-zero)/zero
    d['+24mo'] = (p_24-zero)/zero
    d['+36mo'] = (p_36-zero)/zero
    d['+48mo'] = (p_48-zero)/zero
    return d

    
get_key_stats(exdf)
```

```python
ls = []
for ex in tqdm(r):
    df = ex.pop('df')
    df = df.set_index('ds')
    d = get_key_stats(df)
    d.update({x: ex[x] for x in ex.keys() if x !="df"})
    ls.append(d)
```

```python
ls[0]
```

```python
ndf = pd.DataFrame(data = ls)
ndf.head()
```

```python
ndf.to_csv("../../data/processed/prophet_preds.csv")
```
